```python
#Q1

nums = [1, 2, 3]

nums.append(10)

print(nums)
```

```python
#Q2

order = ["burger", "fries", "coke"]

order.append("ice cream")

order.append("biryani")

order.remove("biryani")

print(order)
```

```python
#Q3

nums = [4, 1, 3, 2]

nums.sort()

print(nums)
```

```python
#Q4

passengers = ["Ali", "Zara", "Ahmed"]

passengers.insert(2, "Sara")

passengers.remove("Zara")

passengers.reverse()

print(passengers)
```

```python
#Q5

text = "hello python world"

modified = text.replace(" ", "_")

words = modified.split("_")

joined = "-".join(words)

print(modified)
print(words)
print(joined)
```

```python
#Q6

word = "Javascript"

print(len(word) > 5)
```

```python
#Q7

product_code = "SKU-12345-XYZ"

print(product_code.find("-"))

print(product_code.count("-"))

number = product_code.replace("SKU-", "").replace("-XYZ", "")
print(number)
```

```python
#Q8

cities = ["Karachi", "Lahore", "Islamabad", "Quetta"]

cities.reverse()

print(cities)
```

```python
#Q9

text = "    Data Analysis with Python is very powerful    "

text = text.strip()

title_text = text.title()

new_text = title_text.replace("Powerful", "Important")

position = new_text.find("Python")

print(title_text)
print(new_text)
print(position)
```

```python
#Q10

data = [12, 45, 2, 67, 34, 89, 23, 90]

data.sort()

print(data)

data.remove(max(data))

print(data)

data.insert(3, 50)

print(data)

data.reverse()

print(data)
```
