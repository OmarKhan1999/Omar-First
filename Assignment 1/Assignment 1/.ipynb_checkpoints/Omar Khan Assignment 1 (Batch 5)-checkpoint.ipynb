```python
#Q3. "I have 22 apples and 33 mangoes. How many fruits do I have in total?" apple = '22' mangoes = 33

apple = '22'
mangoes = 33

# Converting apple from string to integer
total_fruits = int(apple) + mangoes

print("Total fruits:", total_fruits)
```

```python
#Q4. Check the type and convert

num1 = 10
num2 = 5.6
num3 = 4.43
num4 = '23'

# Checking types
print(type(num1))
print(type(num2))
print(type(num3))
print(type(num4))

# Integer to float
num1_float = float(num1)

# Float to integer
num2_int = int(num2)

# Float to string
num3_str = str(num3)

# String to integer
num4_int = int(num4)

#results
print("Integer to float:", num1_float)
print("Float to integer:", num2_int)
print("Float to string:", num3_str)
print("String to integer:", num4_int)
```

```python
#Q6

a = 2
b = 1
c = 6
d = 3

add_ab = a + b
sub_ca = c - a
mul_db = d * b
div_cd = c / d
mod_cd = c % d
square_da = (d + a) ** 2

print("a + b =", add_ab)
print("c - a =", sub_ca)
print("d * b =", mul_db)
print("c / d =", div_cd)
print("c % d =", mod_cd)
print("(d + a)^2 =", square_da)
```
